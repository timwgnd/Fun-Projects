```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simulation with n = 1000 samples

# Generate n = 1000 random points in the unit square [0,1] × [0,1]
n = 1000
x = np.random.rand(n, 2)

# Identify points that satisfy the equation of a quarter-circle
inside = x[np.sqrt(x[:,0]**2 + x[:,1]**2) < 1]

# Estimate π Using Monte Carlo Simulation
# Area of square is 1, area of quarter circle is π/4, fraction of points inside quarter circle is ~ π/4
estimate = 4 * len(inside)/len(x)

print("Estimate of pi:", estimate)

plt.figure(figsize = (4,4))
plt.scatter(x[:,0], x[:,1], s = 0.5, c = "red")
plt.scatter(inside[:,0], inside[:, 1], s = 0.5, c = "blue")
plt.show()
```

```python
n = 10000
estimates = []

for i in range(1, n):
    x = np.random.uniform(-1, 1, (i, 2))
    inside = np.sum(x[:, 0]**2 + x[:, 1]**2 <= 1)
    estimate = 4 * inside / i
    estimates.append(estimate)

errors = [np.abs(np.pi - estimate) for estimate in estimates]

plt.figure(figsize = (10, 5))
plt.subplot(1, 2, 1)
plt.plot(range(1, n), estimates, label="Estimated π", c = "blue")
plt.axhline(y = np.pi, c = "red", linestyle = "--", label = "Actual π")
plt.xlabel("Number of Samples")
plt.ylabel("Estimated π")
plt.legend()
plt.title("Monte Carlo Estimation of π")

plt.subplot(1, 2, 2)
plt.plot(range(1, n), errors, label = "Error", c = "green")
plt.xlabel("Number of Samples")
plt.ylabel("Error |π - estimate|")
plt.legend()
plt.title("Error in Estimation")
```
